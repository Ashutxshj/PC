```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
           row='lunch');#Setting row to make a subplot
#Categorical Plot
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
           col='lunch');#Setting col to make a subplot
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
           col='lunch',row='test preparation course');#Both row,col
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',
           col='lunch',row='race/ethnicity');
```

```python
g=sns.PairGrid(df);
g=g.map_upper(sns.scatterplot) #UPPER
```

```python
g=sns.PairGrid(df);
g=g.map_lower(sns.kdeplot); #LOWER
```

```python
g=sns.PairGrid(df);
g=g.map_diag(sns.kdeplot); #MIDDLE
```

```python
g=sns.PairGrid(df,hue='gender')
g=g.map_diag(sns.kdeplot)
g=g.map_lower(sns.kdeplot)
g=g.map_upper(sns.scatterplot);
```
