```python
import pandas as pd
```

```python
##Conditional Filtering
#Fiter by single/multiple conditions
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df['total_bill']
```

```python
df['total_bill']>40
```

```python
bool_series=df['total_bill']>40
df[bool_series] #Only going to return the rows where the bool_series is true
```

```python
df[df['total_bill']>40] #Same as above
```

```python
df[df['sex'] =='Male']
```

```python
df[df['size']>3]
```

```python
#AND &--- Both conditions need to be true


#OR | --- Either condition is true
```

```python
df[(df['total_bill'] > 30) & (df['sex']=='Male')] #dont use and/or
```

```python
df[(df['total_bill'] > 30) | (df['sex']=='Male')]
```

```python
df[(df['day']=='Sun') | (df['day']=='Sat') | (df['day']=='Fri')]
```

```python
options =['Sat','Sun']
df['day'].isin(options)
```

```python
df[df['day'].isin(options)] #to give data for the days in options only
```
