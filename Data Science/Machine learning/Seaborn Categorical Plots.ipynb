```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df,x='division') #BAR CHART
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
plt.ylim(90,260); #Set limit
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education');
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division'); #Set HUE
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',
           estimator=np.mean,ci='sd',
           hue='division'); #Standard Deviation and Mean

plt.legend(bbox_to_anchor=(1.05,1));
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.boxplot(data=df,y='math score');
```

```python
plt.figure(figsize=(10,4),dpi=500)
sns.boxplot(data=df,y='math score',x='test preparation course');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education',
           hue='test preparation course'); #Adding Categories

plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course',palette='Set2'); 
#Changing x,y

plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education'); 
#Violin Plot
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course',palette='Set2');
plt.legend(bbox_to_anchor=(1.1,0.5));
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course',palette='Set2',
              split=True);#Split

plt.legend(bbox_to_anchor=(1,0.5));
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course',palette='Set2',
              split=True,inner='quartile');#Lines inside it

plt.legend(bbox_to_anchor=(1,0.5));
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course',palette='Set2',
              split=True,inner='stick');

plt.legend(bbox_to_anchor=(1,0.5));
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,
             hue='test preparation course',dodge=True);

#Dodge splits it up
plt.legend(bbox_to_anchor=(1.35,0.5));
```

```python
sns.boxenplot(x='math score',y='test preparation course',data=df,
             hue='gender');
```
