```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df['total_spend']=df['TV']+df['radio']+df['newspaper'] #Linear Equation
df.head()
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales');
```

```python
sns.regplot(data=df,x='total_spend',y='sales'); #Regression Plot
```

```python
X = df['total_spend']
y = df['sales']
```

```python
help(np.polyfit) #y =mx+b
                #y = B1x+B0
```

```python
np.polyfit(X,y,deg=1) #Gives B1,B0
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicated_sales = 0.04868788*potential_spend + 4.24302822
#The Equation
```

```python
plt.plot(potential_spend,predicated_sales); #Graph for Eqn
```

```python
sns.regplot(data=df,x='total_spend',y='sales')
plt.plot(potential_spend,predicated_sales,color='red');
#COMPARISON
```
