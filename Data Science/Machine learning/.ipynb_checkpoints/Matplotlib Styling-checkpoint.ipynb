```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ADDING LEGENDS
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])


ax.plot(x,x,label='X vs X')


ax.plot(x,x**2,label='X vs X^2')

ax.legend() #Adding labels
;
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])


ax.plot(x,x,label='X vs X')


ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc="lower left") #Changing position
;
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])


ax.plot(x,x,label='X vs X')


ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5))#According to add_axes
;
```

```python
#Colors and Styling
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x,color='lightblue');
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#650dbd',label='Purple'); #RGB HEX Colors

ax.plot(x,x+1,color='yellow',label='Yellow')
ax.legend();
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#650dbd',label='Purple',lw=10); #Line width

ax.plot(x,x+1,color='yellow',label='Yellow',linewidth=5)
ax.legend();
```

```python
#possible linestype options '--','-','-.',':','steps'
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#650dbd',label='Purple',lw=10,ls='--',marker='+',
       markeredgecolor='orange',markeredgewidth='10'); #Line width

ax.plot(x,x+1,color='red',label='Red',linewidth=1,ls='-.',marker='o')
ax.legend();
```
