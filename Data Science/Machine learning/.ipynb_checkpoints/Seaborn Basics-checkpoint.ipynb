```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='salary',y='sales',data=df) #BASIC SCATTERPLOT
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df); 
#Seaborn calls matplot at backend
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df); 
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education');
#Dot color #Automatic Legend :D #3 things in a 2D plot
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='salary');#Gradient
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',
               palette='Dark2'); #Palette
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,size='salary');
#Size
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,s=200);#General size
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,size='salary',alpha=0.5);
#Transparency
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,size='salary',
                style='level of education');
#Styles the parameters
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,s=200,
                style='level of education',hue='level of education');
#Style + Hue
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,size='salary',
                style='level of education');

plt.savefig('plot.jpg')
```
