```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots()
```

```python
axes
```

```python
fig,axes=plt.subplots()
axes.plot(x,y); #Can do this manually
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
axes
```

```python
axes.shape
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
axes.shape
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b); #like a matrix with 1st position named 1 so (0,0)=(1,1)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][1].plot(a,b); #MATRIX CONFIRMED
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes: #for all plots
    ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(x,y) 
    
    #For making space
    
plt.tight_layout;
```

```python
#manualy spacing
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=1,hspace=1)#width,height space
;
```

```python

fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=200)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].set_ylabel('Y label 1,0')
axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')#Set title



fig.suptitle('Figure level',fontsize=20)#Figure level 


plt.tight_layout()


fig.savefig('new_subplots.png',bbox_inches='tight') #to save
```
