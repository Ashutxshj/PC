```python
fig=plt.figure()

#LARGE AXES

axes1=fig.add_axes([0,0,1,1])


axes1.plot(a,b)
#SMALL AXES
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
plt.show()
```

```python
fig=plt.figure()

#LARGE AXES

axes1=fig.add_axes([0,0,1,1])


axes1.plot(a,b)
#SMALL AXES
axes2 = fig.add_axes([0.2,0.2,0.1,0.1])#Changing this
plt.show()
```

```python
fig=plt.figure()

#LARGE AXES

axes1=fig.add_axes([0,0,1,1])


axes1.plot(a,b)
#SMALL AXES
axes2 = fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(x,y)
plt.show()
```

```python
fig=plt.figure()

#LARGE AXES

axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("Graph")


axes1.plot(a,b)
#SMALL AXES
axes2 = fig.add_axes([1,1,0.25,0.25])
axes2.plot(a,b)


axes2.set_xlim(1,2)
axes2.set_ylim(0,50) #Super Zoomed In
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title("Zoomed In")


plt.show()
```

```python
#Figure Parameters
```

```python
fig=plt.figure(dpi=200)#To remove blurness
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```

```python
fig=plt.figure(figsize=(12,8),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```

```python
fig=plt.figure(figsize=(10,3),dpi=200) #Changing figure size
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```

```python
#exporting it
fig=plt.figure(figsize=(10,3),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);

fig.savefig('new_figure.png') #Saves super large
```

```python
#exporting it
fig=plt.figure(figsize=(2,2),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);

fig.savefig('new_figure.png')
```

```python
fig=plt.figure(figsize=(2,2),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);

fig.savefig('new_figure.png',bbox_inches='tight')
```
