# SVM = Where a hyperplane distinguishes two labels

```python
df = pd.read_csv('mouse_viral_study.csv')
```

```python
df
```

```python
sns.scatterplot(x='Med_1_mL',y='Med_2_mL',hue='Virus Present',
               data=df)

#HYPERPLANE
x=np.linspace(0,10,100)
m=-1
b=11
y=m*x +b

plt.plot(x,y,'black');
```

## Changing the bias of line

```python
from sklearn.svm import SVC
```

```python
help(SVC)
```

```python
y = df['Virus Present']
```

```python
X = df.drop('Virus Present',axis=1)
```

```python
model = SVC(kernel='linear',C=1000) #Inversely proportional to C
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary #From file
```

```python
plot_svm_boundary(model,X,y)
```

```python
model =SVC(kernel='linear',C=0.05)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model =SVC(kernel='linear',C=0.0005)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

## Do Cross Validation for getting the right C

# Radial Basis Function

```python
model = SVC(kernel='rbf',C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

### Gamma

```python
model = SVC(kernel='rbf',C=1,gamma='auto') #Auto is 0.5
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma=0.9)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma=2)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly',degree=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly',degree=2)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1],'kernel':['linear','rbf']}
```

```python
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```

## 0.01 is the best kernel for it
