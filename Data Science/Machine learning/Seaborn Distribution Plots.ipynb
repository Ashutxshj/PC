```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5); #1D plot 
#SCATTERPLOT
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5);
```

```python
sns.displot(data=df,x='salary');
#HISTOGRAM
```

```python
sns.displot(data=df,x='salary',bins=100); #bins for detail
```

```python
sns.set(style='darkgrid') #Dark background
sns.displot(data=df,x='salary',bins=20);
```

```python
sns.displot(data=df,x='salary',bins=20,color='red',
           edgecolor='blue');#Styling
```

```python
sns.displot(data=df,x='salary',bins=20,color='red',
           edgecolor='blue',linewidth=4);#Thick blue line
```

```python
sns.histplot(data=df,x='salary'); #ANOTHER WAY FOR HISTOGRAM
```

```python
sns.histplot(data=df,x='salary',kde=True); #For curve
```

```python
sns.kdeplot(data=df,x='salary') #Only for KDE
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age');
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30,kde=True);
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100]); #Clip off
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.1);#Bandwidth
```
